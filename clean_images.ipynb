```python
import numpy as np
import glob
import os
import cv2
from skimage.transform import resize
import matplotlib.pyplot as plt
from skimage.color import rgb2grey
```

```python
num_images = len(os.listdir("./downloads/spiral-galaxy"))
X = np.zeros((num_images, 200, 200, 3))
```

```python
for i, image_name in enumerate(os.listdir("./downloads/spiral-galaxy")):
    print(f"#: {i}, image_name: {image_name}")
    img = cv2.imread("./downloads/spiral-galaxy/" + image_name, cv2.IMREAD_COLOR)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    img = resize(img, (200, 200, 3), anti_aliasing=True)
    X[i, :, :, :] = img
```

```python
np.save("./data/images_clean-spiral.npy", X)
```
